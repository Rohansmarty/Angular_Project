```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Read an Excel file
df = pd.read_csv('house_price_dataset_original_v2_cleaned.csv')

# Access data in the DataFrame
print(df.head())  # Display the first few rows of the dataset
```

```python
df
```

```python
# Splitting data
X = df[['land_size_sqm']]
y = df['house_size_sqm']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Training the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# You can print the coefficients to understand the relationship
print('Coefficients:', regressor.coef_)
print('Intercept:', regressor.intercept_)
print('score:', regressor.score(X_train, y_train))
```

```python
# Predicting
y_pred = regressor.predict(X_test)

# Visualization
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('land_size_sqm vs house_size_sqm')
plt.xlabel('land_size_sqm')
plt.ylabel('house_size_sqm')
plt.show()
```

```python
# Gradient descent for Simple Linear Regression

import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
df = pd.read_csv('house_price_dataset_original_v2_cleaned.csv')

def normalize(feature):
    """Standardize the feature using Z-score normalization."""
    return (feature - np.mean(feature)) / np.std(feature)

# Normalize data
X = normalize(np.array(df['land_size_sqm']))
y_normalized = normalize(np.array(df['house_size_sqm']))

# Hyperparameters
alpha = 0.01
num_iterations = 1000

# Initialization
m = len(df['land_size_sqm'])
theta0 = 0  # intercept
theta1 = 0  # slope

# Gradient Descent
for _ in range(num_iterations):
    # Predictions
    y_pred = theta0 + theta1 * X
    
    # Derivatives
    dtheta0 = (1/m) * sum(y_pred - y_normalized)
    dtheta1 = (1/m) * sum((y_pred - y_normalized) * X)
    
    # Update thetas
    theta0 = theta0 - alpha * dtheta0
    theta1 = theta1 - alpha * dtheta1

# Visualization
plt.scatter(X, y_normalized, color='red')
plt.plot(X, theta0 + theta1 * X, color='blue')
plt.title('Normalized Housing land_size_sqm vs Normalized house_size_sqm')
plt.xlabel('Normalized land_size_sqm')
plt.ylabel('Normalized house_size_sqm')
plt.show()
```

```python
# Cost function for Simple linear Regression
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
df = pd.read_csv('house_price_dataset_original_v2_cleaned.csv')

def normalize(feature):
    """Standardize the feature using Z-score normalization."""
    return (feature - np.mean(feature)) / np.std(feature)

# Normalize data
X = normalize(np.array(df['land_size_sqm']))
y_normalized = normalize(np.array(df['house_size_sqm']))


# Hyperparameters
alpha = 0.01
num_iterations = 1400

# Initialization
m = len(df['land_size_sqm'])
theta0 = 0  # intercept
theta1 = 0  # slope

# Cost history to store MSE values for each iteration
cost_history = []

# Gradient Descent
for _ in range(num_iterations):
    y_pred = theta0 + theta1 * X
    cost = (1/m) * sum((y_pred - y_normalized)**2)
    cost_history.append(cost)
    
    dtheta0 = (1/m) * sum(y_pred - y_normalized)
    dtheta1 = (1/m) * sum((y_pred - y_normalized) * X)
    
    theta0 = theta0 - alpha * dtheta0
    theta1 = theta1 - alpha * dtheta1

plt.plot(cost_history)
plt.title('Cost Function Over Iterations (Simple Regression)')
plt.xlabel('Iterations')
plt.ylabel('Cost (MSE)')
plt.show()
```
